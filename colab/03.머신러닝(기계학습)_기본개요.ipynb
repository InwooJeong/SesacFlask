# 개요

- 머신러닝 사용 이유
  - AI > 머신러닝 > 딥러닝
  - 사람의 지성/직관으로 해결할 수 없는 문제들을 해결하기 위해 새로운 개념이 도입된 분야
  - 프로그래밍 관점

  ```
    # 기존
    # 로직, 알고리즘을 통해 해결
    def sum(x,y):
      return x+y

    # 복잡한 문제 -> 기존 방식으로 해결이 불가
    def defect_photo(img):
      if ...:  # 동물이고
        if ...:  # 고양이 ...
        ...
      # 이러한 방식으로는 한계
    
    # 제안 -> 대량의 데이터를 이용, 학습을 수행 -> 특정 결과에 대한 특징을 이해 -> 한 번도 만나지 않은 데이터를 받을 시 판별/판단/예측
    # 머신러닝(알고리즘 활용), 딥러닝(인공신경망 활용)
    # 머신러닝 => 정형, 반정형 데이터 특화
    # 딥러닝 => 비정형 데이터 특화
    # 프로그래밍 관점 => 그냥 라이브러리
  ```
